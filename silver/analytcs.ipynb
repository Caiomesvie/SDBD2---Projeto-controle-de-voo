### 1.1. Contexto e Fontes de Dados

Os nossos dados foram obtidos a partis do portal de [Acesso a Informação](https://www.gov.br/inep/pt-br/acesso-a-informacao) do proprio governo, que fica dentro do sitema do [Inep](https://www.gov.br/inep/pt-br), onde segue a seguinte navegação ate chegar nos dados: [**Dados Abertos** > **Microdados** > **Enem**](https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem) onde o acesso aos dados são de acesso livre a todos.


- **Fonte Principal:** [Enem](https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem).

### Sobre o Enem
O Enem [(Exame Nacional do Ensino Médio)](https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem) é a principal avaliação educacional do Brasil. Ele serve como a porta de entrada para o ensino superior, sendo utilizado para ingressar em universidades públicas (via Sisu), conseguir bolsas em faculdades particulares (ProUni) e obter financiamento estudantil (Fies).

**Em resumo:** É a prova oficial do governo que avalia o aprendizado escolar e define quem entra na faculdade.
